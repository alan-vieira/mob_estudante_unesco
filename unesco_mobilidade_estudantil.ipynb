###Inbound marketing

Inbound marketing ou marketing de atra√ß√£o √© uma metodologia de neg√≥cios na qual uma empresa se promove por meio de conte√∫dos publicados na internet, em blogs, v√≠deos, ebooks, not√≠cias e outras formas de marketing de conte√∫do.

No Inbound Marketing √© o cliente que busca pela empresa e n√£o o contr√°rio. Tem como caracter√≠sticas uma atua√ß√£o mais passiva na gera√ß√£o de leads e vendas.

```python
# Marketing de atra√ß√£o üß≤ (s√©rie hist√≥rica) 
df.groupby('country')['total_inbound'].sum().sort_values(ascending = False).head(10)
```

###Outbound marketing

Outbound marketing √© o marketing tradicional, onde a institui√ß√£o de ensino vai em busca dos estudantes e oferece produtos e servi√ßos mesmo que estes n√£o tenham demonstrado interesse direto.

Tem como caracter√≠sticas uma atua√ß√£o totalmente ativa na gera√ß√£o de leads e vendas.

```python
# Marketing tradicional üì£ (s√©rie hist√≥rica)
df.groupby('country')['total_outbound'].sum().sort_values(ascending = False).head(10)
```
